# Анализ на температурни данни и енергопотребление

## Въведение
Този проект има за цел да анализира връзката между температурните данни и енергопотреблението през 30-дневен период. Данните съдържат липсващи стойности за температурата в определени дни, което изисква използването на интерполационни методи за попълването им. Освен това, ще изследвам статистическите характеристики на данните и връзката между температурата и енергопотреблението.

## Описание на задачата
- **Липсващи стойности**: Температурните данни имат липсващи стойности за дни **5, 12, 19, 26 и 29**. Те ще бъдат попълнени с помощта на интерполационни методи.
- **Сравнение на интерполационни методи**: Ще сравня **интерполационния полином на Лагранж** и **Сплайн интерполацията**.
- **Статистически анализ**: Ще изчисля **средна стойност**, **дисперсия**, **стандартно отклонение**, **коефициент на асиметрия** и **ексцес** за температурните данни и енергопотреблението.
- **Ковариация и корелация**: Ще изследвам връзката между интерполираната температура и енергопотреблението.
- **Визуализация**: Ще представя графично резултатите от интерполациите и зависимостта между температурата и енергопотреблението.

---

## Решение

### 1. Попълване на липсващите стойности

Липсващите стойности за температурата са попълнени с два метода:

- **Интерполационен полином на Лагранж**: Този метод използва полином, който минава точно през известните точки.
- **Сплайн интерполация**: Този метод използва кубични функции за по-гладко приближение.

Резултатите от двете интерполации са показани в таблицата по-долу.

### 2. Сравнение на интерполационните методи

Сравних двата метода, използвайки **средноквадратичната грешка (MSE)**:

- **Лагранж**: MSE = 0.000000
- **Сплайн**: MSE = 0.000000

И двата метода дават точно съвпадение с известните точки, но **Сплайн интерполацията** обикновено предлага по-гладки криви между точките.

Резултатите показват **силна отрицателна корелация** между температурата и енергопотреблението. Това означава, че с увеличаване на температурата, **енергопотреблението намалява** и обратно.

## 5. Визуализации

Графиките показват:

- **Интерполационните методи**: Сравнение между Лагранж и сплайн интерполациите. Сплайн интерполацията дава по-гладка крива.
- **Зависимост между температура и енергопотребление**: Ясно се вижда линейна зависимост с отрицателен наклон, което потвърждава силната отрицателна корелация.

---

## Заключение

Анализът показа, че **Сплайн интерполацията** е подходящ метод за попълване на липсващите температурни стойности. Съществува **силна отрицателна връзка** между температурата и енергопотреблението, което може да се използва за **прогнозиране на енергийните нужди** в бъдеще.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange, interp1d
from sklearn.metrics import mean_squared_error

days = np.arange(1, 31)
temperature = [20.30, 22.30, 23.67, 25.98, np.nan, 28.46, 29.21, 30.05, 29.80, 29.56,
               28.96, np.nan, 25.68, 24.22, 21.98, 20.05, 17.72, 16.23, np.nan, 12.72,
               11.24, 10.69, 9.80, 10.15, 10.34, np.nan, 12.62, 14.02, np.nan, 18.12]

energy = [110.20, 104.20, 104.66, 97.54, 96.64, 91.08, 92.08, 92.10, 89.40, 91.88,
          91.58, 93.34, 99.14, 100.76, 107.24, 109.40, 114.86, 116.04, 122.66, 123.56,
          129.02, 127.92, 130.70, 128.50, 130.12, 126.42, 125.76, 121.26, 119.94, 112.26]

df = pd.DataFrame({'Day': days, 'Temperature': temperature, 'Energy': energy})

def local_lagrange_interpolation(x_known, y_known, x_missing, k=5):
    y_interp = []

    for x in x_missing:
        distances = np.abs(x_known - x)
        nearest_indices = np.argsort(distances)[:k]
        x_nearest = x_known[nearest_indices]
        y_nearest = y_known[nearest_indices]
        poly = lagrange(x_nearest, y_nearest)
        y_interp.append(poly(x))

    return np.array(y_interp)

known_days = df[df['Temperature'].notna()]['Day'].values
known_temp = df[df['Temperature'].notna()]['Temperature'].values
missing_days = df[df['Temperature'].isna()]['Day'].values

lagrange_temp = local_lagrange_interpolation(known_days, known_temp, missing_days, k=5)

for day, temp in zip(missing_days, lagrange_temp):
    df.loc[df['Day'] == day, 'Lagrange_Temp'] = temp

spline = interp1d(known_days, known_temp, kind='cubic', fill_value='extrapolate')
spline_temp = spline(missing_days)

df['Spline_Temp'] = df['Temperature']
for day, temp in zip(missing_days, spline_temp):
    df.loc[df['Day'] == day, 'Spline_Temp'] = temp

lagrange_temp_known = local_lagrange_interpolation(known_days, known_temp, known_days, k=5)
mse_lagrange = mean_squared_error(known_temp, lagrange_temp_known)
mse_spline = mean_squared_error(known_temp, spline(known_days))

print("Сравнение на интерполационните методи:")
print(f"Средноквадратична грешка за Лагранж: {mse_lagrange:.6f}")
print(f"Средноквадратична грешка за Сплайн: {mse_spline:.6f}\n")

temp_stats = {
    'Средна стойност': df['Spline_Temp'].mean(),
    'Дисперсия': df['Spline_Temp'].var(),
    'Стандартно отклонение': df['Spline_Temp'].std(),
    'Коефициент на асиметрия': df['Spline_Temp'].skew(),
    'Ексцес': df['Spline_Temp'].kurtosis()
}

print("Статистически характеристики за температурите:")
for key, value in temp_stats.items():
    print(f"{key}: {value:.4f}")

energy_stats = {
    'Средна стойност': df['Energy'].mean(),
    'Дисперсия': df['Energy'].var(),
    'Стандартно отклонение': df['Energy'].std(),
    'Коефициент на асиметрия': df['Energy'].skew(),
    'Ексцес': df['Energy'].kurtosis()
}

print("\nСтатистически характеристики за енергопотреблението:")
for key, value in energy_stats.items():
    print(f"{key}: {value:.4f}")

covariance = np.cov(df['Spline_Temp'], df['Energy'])[0, 1]
correlation = df['Spline_Temp'].corr(df['Energy'])

print("\nВръзка между температурата и енергопотреблението:")
print(f"Ковариация: {covariance:.4f}")
print(f"Коефициент на корелация: {correlation:.4f}")

plt.figure(figsize=(12, 6))
plt.scatter(df['Day'], df['Temperature'], label='Измерени данни', color='black', s=50)
plt.plot(df['Day'], df['Lagrange_Temp'], label='Локален Лагранж', linestyle='--', linewidth=2)
plt.plot(df['Day'], df['Spline_Temp'], label='Сплайн', linestyle='-', linewidth=2)
plt.xlabel('Ден', fontsize=12)
plt.ylabel('Температура (°C)', fontsize=12)
plt.title('Сравнение на интерполационните методи', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(df['Spline_Temp'], df['Energy'], color='blue', s=60, alpha=0.7)
plt.xlabel('Температура (°C)', fontsize=12)
plt.ylabel('Енергопотребление (ед.)', fontsize=12)
plt.title('Зависимост на енергопотреблението от температурата', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

z = np.polyfit(df['Spline_Temp'], df['Energy'], 1)
p = np.poly1d(z)
plt.plot(df['Spline_Temp'], p(df['Spline_Temp']), "r--", linewidth=2)
plt.show()

print("\nТаблица с интерполираните стойности:")
display(df[['Day', 'Temperature', 'Lagrange_Temp', 'Spline_Temp', 'Energy']])
```
